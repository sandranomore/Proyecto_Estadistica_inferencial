```python
# importamos las librerías que necesitamos

# Tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd
import numpy as np

# Imputación de nulos usando métodos avanzados estadísticos
# -----------------------------------------------------------------------
from sklearn.impute import SimpleImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.impute import KNNImputer

# Librerías de visualización
# -----------------------------------------------------------------------
import seaborn as sns
import matplotlib.pyplot as plt
# Configuración
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

```python
df = pd.read_csv("world_data_notnull.csv")
```

```python
# 1 ¿Cómo se distribuye la tasa de natalidad en el conjunto de datos? ¿Hay algún patrón evidente?

# Recurrimos a un histograma

# creamos nuestra primera gráfica para saber cómo está distribuida la tasa de natalidad entre el conjunto de datos
# para ello usamos el método 'sns.histplot', en este caso le hemos cambiado el color por defecto usando el parámetro 'color' 
sns.histplot(x = "birth_rate", data = df, color = "blue")

# cambiamos los nombres de los ejes
plt.title("Distribución tasa natalida")

plt.xlabel("tasa de natalidad");

"""La conclusión es que a menor valor mayor concentración de datos"""
```

```python
# 2. ¿Existe una relación aparente entre la tasa de mortalidad infantil y la esperanza de vida? Proporciona una breve descripción de los patrones que observes.

# # Recurrimos a un gráfico de regresión

# creamos una gráfica básica usando 'sns.scatterplot()' 
# como ya lo hemos aprendido, vamos a cambiar el nombre de los ejes usando los métodos 'plt.xlabel()'  y 'plt.ylabel()' 
sns.scatterplot(x = "infant_mortality", 
                y = "life_expectancy", 
                data = df)

# cambiamos los nombres de los ejes
plt.title("Relación entre mortalidad infantil y esperanza de vida")

# cambiamos los nombres de los ejes
plt.xlabel("tasa de mortalidad infantil")
plt.ylabel("esperanza de vida");

# Las variables tienen una corelación negativa. A menor mortalidad infartil mayor esperanza de vida y viceversa. 
```

```python
# 3 ¿Cuál es el continente con la mayor cantidad de países en el conjunto de datos? Proporciona el porcentaje de países en ese continente.

num_country = df.groupby("continent")["country"].count().reset_index()
num_country
```

```python
# 4 ¿Cuál es la mediana de la esperanza de vida en Europa? Proporciona un rango intercuartil aproximado y verifica si existen valores atípicos.

# comenzamos empezando estableciendo nuestra gráfica y indicando el número de filas y columnas que queremos
fig, axes = plt.subplots(nrows = 1, ncols = 4, figsize = (20,5))

sns.boxplot(x = "life_expectancy", 
            y = "continent", 
            data = df,  
            ax = axes[0],
            palette= "mako");
```
