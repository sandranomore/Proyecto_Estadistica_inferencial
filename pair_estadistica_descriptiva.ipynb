
Calcula la tasa promedio de natalidad (birth_rate) para todos los países en el conjunto de datos.


```python
tasa_promedio_natalidad = df["birth_rate"].mean().round(2)
print("Tasa promedio natalidad:", tasa_promedio_natalidad)
```


Realiza un análisis de la mortalidad infantil (infant_mortality) y clasifica los países en tres categorías: baja, moderada y alta mortalidad infantil.


```python
# Función para clasificar la mortalidad infantil
def clasificar_mortalidad(mortalidad):
    if mortalidad <= 10:
        return 'baja'
    elif mortalidad <= 30:
        return 'moderada'
    else:
        return 'alta'

# Aplicar la función al DataFrame
df['infant_mortality_category'] = df['infant_mortality'].apply(clasificar_mortalidad)

# Contar la cantidad de países en cada categoría
mortalidad_infantil_clasificacion = df['infant_mortality_category'].value_counts()

# Mostrar la clasificación de mortalidad infantil
print("Clasificación de mortalidad infantil:\n", mortalidad_infantil_clasificacion)
```


Calcula la correlación entre las variables de "fertility_rate" y "life_expectancy". Interpreta la relación entre estas dos variables. ¿Qué tipo de correlación hay? Explica la razón de esta relación
