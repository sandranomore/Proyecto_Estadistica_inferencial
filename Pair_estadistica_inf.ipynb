```python
# importamos las librerías que necesitamos

# Tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd
import numpy as np

# Visualización
# ------------------------------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns

# Evaluar linealidad de las relaciones entre las variables
# y la distribución de las variables
# ------------------------------------------------------------------------------
#import scipy.stats as stats
import scipy.stats as stats
from scipy.stats import shapiro, poisson, chisquare, expon, kstest


# Configuración
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames

# Gestión de los warnings
# -----------------------------------------------------------------------
import warnings
warnings.filterwarnings("ignore")
```

```python

# Leer el archivo con la codificación detectada
df = pd.read_csv("world_data_notnull.csv", index_col=0)
df.head()
```

```python
df.isnull().sum()
```

```python
# Nivel de confianza esperanza de vida al 95%
esperanza_vida = df['life_expectancy']
esperanza_vida
```

```python
# Nivel de confianza esperanza de vida al 95%
confianza = 0.95
n = len(esperanza_vida)
media_muestral = np.mean(esperanza_vida)
desviacion_estandar_muestral = np.std(esperanza_vida, ddof=1)
error_estandar = desviacion_estandar_muestral / np.sqrt(n)# Valor crítico (distribución t de Student)
t_critico = stats.t.ppf((1 + confianza) / 2, df=n-1)# Margen de error
margen_error = t_critico * error_estandar# Intervalo de confianza
limite_inferior = media_muestral - margen_error
limite_superior = media_muestral + margen_error# Resultados
print(f"Media Muestral: {np.round(media_muestral, 2)}")
print(f"Error Estándar: {np.round(error_estandar, 2)}")
print(f"Nivel de Confianza: {confianza}")
print(f"Valor Crítico: {np.round(t_critico, 2)}")
print(f"Intervalo de Confianza: ({np.round(limite_inferior, 2)}, {np.round(limite_superior, 2)})")
```

```python
# Nivel de confianza PIB al 90%
confianza = 0.90
n = len(esperanza_vida)
media_muestral = np.mean(esperanza_vida)
desviacion_estandar_muestral = np.std(esperanza_vida, ddof=1)
error_estandar = desviacion_estandar_muestral / np.sqrt(n)# Valor crítico (distribución t de Student)
t_critico = stats.t.ppf((1 + confianza) / 2, df=n-1)# Margen de error
margen_error = t_critico * error_estandar# Intervalo de confianza
limite_inferior = media_muestral - margen_error
limite_superior = media_muestral + margen_error# Resultados
print(f"Media Muestral: {np.round(media_muestral, 2)}")
print(f"Error Estándar: {np.round(error_estandar, 2)}")
print(f"Nivel de Confianza: {confianza}")
print(f"Valor Crítico: {np.round(t_critico, 2)}")
print(f"Intervalo de Confianza: ({np.round(limite_inferior, 2)}, {np.round(limite_superior, 2)})")
```
